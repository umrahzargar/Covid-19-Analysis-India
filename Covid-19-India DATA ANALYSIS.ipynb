```python
import pandas as pd
import numpy as np
import matplotlib 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data["Updated On"] = pd.to_datetime(data["Updated On"],dayfirst=True)
data.info()
```

```python
data.isnull().sum()
```

```python
data.fillna(0,inplace=True)
```

```python
data.isnull().sum()
```

```python
data.head()
```

```python
data["State"].unique()
```

```python
data.drop(data[data["State"] == "India"].index, inplace=True)
```

```python
data["State"].unique()
```

```python
data=data.reset_index(drop=True)
```

```python
data.head()
```

```python
vaccinated = data.groupby("State")[["Total Individuals Vaccinated"]].max().sort_values("Total Individuals Vaccinated",ascending=False)
plt.figure(figsize=(18,10))
plt.title("Total Individuals done in states of India")

sns.barplot(x=vaccinated["Total Individuals Vaccinated"], y=vaccinated.index);
```

```python
doses = data.groupby("State")[["First Dose Administered","Second Dose Administered"]].max()
```

```python
plt.figure(figsize=(18,10))

xpos = np.arange(len(doses.index))
plt.xticks(xpos,doses.index,rotation=89)


plt.bar(xpos-0.2,doses["First Dose Administered"]/1e7, label="Dose 1",width=0.4)
plt.bar(xpos+0.2,doses["Second Dose Administered"]/1e7, label="Dose 2",width=0.4)


plt.xlabel("States",fontweight="bold")
plt.ylabel("Number of doses in crores",fontweight="bold")
plt.title("Dose 1 and Dose 2 vaccination coverage",fontweight="bold")
plt.legend();
```

```python
fig = plt.figure(figsize=(20,10))
df_t=data.groupby(['State']).agg({'Total Doses Administered':np.mean})
df_t1=df_t.nlargest(10, ['Total Doses Administered'])
df_t1.groupby(["State"]).sum()["Total Doses Administered"].plot(kind='pie',rot=0,autopct='%1.0f%%', textprops={'fontsize': 20},explode=[0.01,0.01,0.05,0.01,0.01,0.01,0.01,0.05,0.01,0.01],shadow=True)
plt.title("Total Doses Administered wrt States",fontsize=15)
plt.legend()
```

```python
second_dose_jnk= data[data['State']=="Jammu and Kashmir"]
second_dose_jnk
```

```python
fig = plt.figure(figsize=(20,10))
df_t=second_dose_jnk.groupby(['Sessions']).agg({'Male (Doses Administered)':np.mean})
df_t1=df_t.nlargest(10, ['Male (Doses Administered)'])
df_t1.groupby(["Sessions"]).sum()["Male (Doses Administered)"].plot(kind='pie',rot=0,autopct='%1.0f%%', textprops={'fontsize': 20},explode=[0.01,0.01,0.05,0.01,0.01,0.01,0.01,0.05,0.01,0.01],shadow=True)
plt.title("Total sessions w.r.t Male gender in Jammu and Kashmir",fontsize=15)
plt.legend()
```
